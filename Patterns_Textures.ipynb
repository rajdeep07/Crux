```python
# Reading the extracted images from Vogue

df = pd.read_csv('/Users/nitsy/Desktop/Personal/Current_InProgress/method_data_science/Crux-master/Exported_Images_v2', usecols=[1,2,3,4])
```

```python
# My API Key
app = ClarifaiApp(api_key='d3507df257b840f783d75d5111866eae')

# get the general model
model = app.models.get("Textures & Patterns")

data_final = pd.DataFrame()

#Iterating through each row in the dataframe and passing the imageurl to the Predict API
#Then extracting the texture and its value from the response

for id in df['id']:
    url_img = df.loc[df['id'] == id, 'imageurl'].iloc[0]
    image = ClImage(url=str(url_img))
    response = model.predict([image])
    for i in range(0,len(response['outputs'][0]['data']['concepts']),1):
        p={'id': id, 'texture': str(response['outputs'][0]['data']['concepts'][i]['name']),'value': response['outputs'][0]['data']['concepts'][i]['value']}
        data_final = data_final.append(p, ignore_index=True)
```
